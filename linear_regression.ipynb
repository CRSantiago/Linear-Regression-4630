## 3 - Single Variable Nonlinear Regression ##


Develop a regression model, estimate coefficients with data, and derive the relationship. 

**Tasks:**
1. Establish a relationship between Y and X with a cubic function.
2. Compute MSE loss with observation-prediction pairs. 
3. Implement **Gradient Descent (GD)** to achieve optimal solution with the learning rate of **0.000001 (1e-6)** and **10000 (1e4)** epochs.
4. Print out the optimal solution at final step. 




```python
a = 0   # cubic
b = 0   # quadratic
c = 0   # linear
d = 0   # constant

L = 0.000001  # learning rate
epochs = 10000

n = len(X)

for i in range(epochs):
  Y_pred = a*(X**3) + b*(X**2) + c*X + d  # current prediced value of Y

  D_a = (-2/n) * sum(X**3 * (Y - Y_pred))   # Dervative wrt a
  D_b = (-2/n) * sum(X**2 * (Y - Y_pred))   # Dervative wrt b
  D_c = (-2/n) * sum(X * (Y - Y_pred))      # Dervative wrt c
  D_d = (-2/n) * sum((Y - Y_pred))          # Dervative wrt d

  a = a - (L * D_a)
  b = b - (L * D_b)
  c = c - (L * D_c)
  d = d - (L * D_d)
print(a,b,c,d)
```


## 5.2 Linear Regression

Develop a regression model, estimate coefficients with data, and derive the relationship. 

**Tasks:**
1. Establish a linear function to describe the relationship among Y, X1, and X2.
2. Compute MSE loss with observation-prediction pairs.
3. Implement **Gradient Descent (GD)** to achieve optimal solution with the learning rate of **0.001 (1e-3)** and **10000 (1e4)** epochs.
4. Print out the optimal solution at final step. 


```python
m1 = 0   # X1 coefficient
m2 = 0   # X2 coefficient
m3 = 0   # constant

L = 0.001  # learning rate
epochs = 10000

n = len(X1)

count = 0

for i in range(epochs):
  Y_pred = m1*X1 + m2*X2 + m3  # current prediced value of Y

  D_m1 = (-2/n) * sum(X1 * (Y - Y_pred))   # Dervative wrt m1
  D_m2 = (-2/n) * sum(X2 * (Y - Y_pred))      # Dervative wrt m2
  D_m3 = (-2/n) * sum((Y - Y_pred))          # Dervative wrt m3

  m1 = m1 - (L * D_m1)
  m2 = m2 - (L * D_m2)
  m3 = m3 - (L * D_m3)
  
  # for debugging
  # if(count <= 5):
  #   print(m1, m2, m3)
  #   count += 1
print(m1, m2,m3)
```
